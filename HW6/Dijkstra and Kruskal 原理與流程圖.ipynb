Dijkstra 演算法 

是從一個頂點到其餘各頂點的最短路徑演算法，解決的是有向圖中最短路徑問題。 其基本原理是：每次新擴展一個距離最短的點，更新與其相鄰的點的距離。
1.	此演算法是以某一個節點為起點，計算該節點到其他節點的最短距離
2.	通過走訪不同的節點，可以再找出起點到其他節點間更短的路徑



Kruskal 演算法

最小生成樹(Minimum Spanning Tree)
在無向圖有權重的連通圖中找尋可以連接所有點的邊且不形成循環，且這些邊的權重和最小，可以連通所有點且不形成循環，一定會形成樹，這樣的問題稱作最小生成樹(Minimum Spanning Tree)。
介紹Kruskal的最小生成樹演算法，因為這個演算法較容易實作，Kruskal演算法由最小的邊出發，找出最小不形成循環的邊，直到邊的個數為點的個數少1，就找到最小生成樹。
1.	這個演算法是利用計算每個節點到結點之間的權重來做最短路徑
2.	由最小的節點開始尋找依序向上，若形成為圈則捨棄該路徑
3.	要做到的是最後節點間權重相加為最小值
