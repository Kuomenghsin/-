# BFS廣度優先搜尋 (Breadth-First Search)

廣度優先搜尋法（Breadth-First Search, BFS）是一種樹（Tree）或圖（Graph）資料結構的搜索演算法，從圖的某一節點(vertex, node) 開始走訪，接著走訪此一節點所有相鄰且未拜訪過的節點，才進一步走訪下一層級的節點，因為它是每一個層級都走房完之後才會在進入下一個層級去進行走訪的，所以廣度優先搜尋類似階層的概念去做實作與應用。

#  DFS 深度優先搜尋 (Depth-First Search)

深度優先搜尋法（Depth-First Search, DFS）也是一種樹（Tree）或圖（Graph）資料結構的搜索演算法，從圖的某一節點(vertex, node) 開始走訪，接著走訪此一節點下方的每一個節點，在下方沒有任何節點之後才會又跳回原階層的另一個節點，又跑完該分支下方所有節點之後，又到其他階層去做相同的動作，這種做法比較盲目沒有盡頭，感覺要花比較多的時間才能夠搜尋到自己想要的東西。

# ＢＦＳ和ＤＦＳ之比較

1. BFS是利用queue來做最短路徑，DFS是利用stack來做最短路徑
2. 在目標比較在目標比較近的時候BFS比較好用，在目標比較遠的時候DFS比較好用
3. BFS會消耗較多的記憶體，但是比較快
4. DFS會消耗較少記憶體，可是查找方向固定，所以比較浪費時間
5. BFS在尋找上偏向一層一層去做搜索，比較有效率，但遇到位置在某一節點的最末端的時候反而會比較浪費時間
6. DFS在尋找上會一次性探究到每一個分支的最末端，這個做法比較盲目沒有效率
